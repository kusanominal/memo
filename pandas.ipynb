### Series

```python
# ラベル(重複可)と値で定義
ser = pd.Series([1, 2, 3], index=['a', 'b', 'c'])
ser
```

```python
# ラベルを指定しないと0-originの数列
pd.Series([1,2,3])
```

```python
# locでラベルからデータ取得(省略可能)
ser.loc['b']
ser['b']
```

```python
# スライシング
# リストのスライシングと違い終了点を含む
ser.loc['b':'c']
```

```python
# 複数要素の取り出し
ser.loc[['a', 'c']]
```

```python
# ラベルでなく整数値でのインデックス指定
ser.iloc[0]
```

```python
# ilocのスライスはリストと同じ
ser.iloc[1:2]
```

```python
# 真偽値による取得
ser.loc[[True, False, True]]
ser.iloc[[True, False, True]]
```

```python
# スカラーと比較すると真偽値のリストを返す
ser != 2
```

```python
# これを利用してインデックス条件を指定
ser.loc[ser != 2]
```

### DataFrame

```python
df = pd.DataFrame([[1, 10, 100], [2, 20, 200], [3, 30, 300]], index=['r1', 'r2', 'r3'], columns=['c1', 'c2', 'c3'])
df
```

```python
df.loc['r2', 'c2']
```

```python
# :ですべての行や列を指定
df.loc['r2', :]
```

```python
# スライシング
df.loc[['r1', 'r3'], 'c2':'c3']
```

```python
# ilocで位置を指定した抜き出し
df.iloc[1:3, [0, 2]]
```

```python
# locやiloc抜きでは列が抜き出される
df['c2']
```

```python
# 比較演算で真偽値が帰る
df > 10
```

```python
# 真偽値による抽出
df.loc[df['c2'] > 10]
```

```python
# 条件を複数にする場合は & | を使う
df.loc[(df['c1'] > 1) & (df['c3'] < 300)]
```

### データ処理

```python
df = pd.read_csv(anime_csv)
df.loc[df['episodes'] == 'Unknown'].head()
```

```python
# whereで条件抽出すると該当しないデータをNanで埋める
df.where(df['rating'] < 9.2).head()
```

```python
# 値の変更
df.loc[0, 'anime_id'] = 12345
df.head()
```

```python
# 複数の値書き換え
df.loc[df['episodes'] == 'Unknown', 'episodes'] = np.nan
df.loc[pd.isnull(df['episodes'])].head()
```

```python
# 欠損値の除外
# 非破壊的操作
# 破壊的操作をする場合キーワードにinplace=Trueとする
df.dropna().loc[70:].head()
```

```python
# 型変換
# astype も非破壊的操作
df['episodes'].astype(np.int64)
df.astype({'episodes':np.int64, 'rating':np.float64})
```

```python
# sort
# 非破壊的操作
df.sort_values('rating', ascending=False).head()
```

```python
# map: series
# apply: dataframeの列または行(axis=1)
# applymap: dataframeの各値
df.apply(len)
```

```python
df.apply(len, axis=1)
```

```python
# applyで渡される関数には引数渡しの際Seriesが渡されるので、ラベルを指定できる
df = df.dropna()
df.apply(lambda x: len(x['name']) + len(x['genre']), axis=1).head()
```
